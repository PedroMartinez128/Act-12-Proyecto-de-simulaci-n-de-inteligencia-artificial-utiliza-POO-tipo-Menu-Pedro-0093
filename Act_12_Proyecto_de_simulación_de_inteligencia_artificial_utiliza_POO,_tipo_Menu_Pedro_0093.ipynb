#### class investigador

```python
class Investigador:
    # --- CONSTRUCTOR DE LA CLASE ---
    def __init__(self, linea=None, publicaciones=None, universidad=None, becas=None, sni=None, laboratorio=None):
        # Si pasamos parámetros, creamos un investigador individual
        self.linea = linea
        self.publicaciones = publicaciones
        self.universidad = universidad
        self.becas = becas
        self.sni = sni
        self.laboratorio = laboratorio

        # Atributo interno: Aquí vive la lista, 100% ADENTRO del objeto del sistema
        self.sistema_ia_investigadores = []

    # --- MÉTODO: AGREGAR ---
    def agregar_investigador(self):
        print("\n--- 🧪 REGISTRAR NUEVO INVESTIGADOR DE IA ---")
        linea = input("Línea de investigación (ej. Redes Neuronales, Robótica): ")
        publicaciones = input("Número de publicaciones científicas: ")
        universidad = input("Universidad de procedencia: ")
        becas = input("Becas obtenidas (ej. Conahcyt, MIT Fellowship): ")
        sni = input("Nivel de SNI (Candidato, I, II, III o Nivel 0 si no tiene): ")
        laboratorio = input("Nombre del laboratorio asignado: ")

        # Creamos el objeto con los datos recibidos
        nuevo_investigador = Investigador(linea, publicaciones, universidad, becas, sni, laboratorio)
        # Lo guardamos en la lista interna usando 'self'
        self.sistema_ia_investigadores.append(nuevo_investigador)
        print("\n✅ ¡Investigador agregado con éxito al sistema de IA!")

    # --- MÉTODO: LISTAR ---
    def listar_investigadores(self):
        print("\n" + "="*115)
        print(f"{'ID':<4} | {'Línea de Investigación':<25} | {'Pubs':<6} | {'Universidad':<20} | {'Becas':<20} | {'SNI':<10} | {'Laboratorio':<15}")
        print("="*115)

        if len(self.sistema_ia_investigadores) == 0:
            print(f"{'NO HAY INVESTIGADORES REGISTRADOS AÚN':^115}")
        else:
            for indice, inv in enumerate(self.sistema_ia_investigadores):
                print(f"{indice + 1:<4} | {inv.linea:<25} | {inv.publicaciones:<6} | {inv.universidad:<20} | {inv.becas:<20} | {inv.sni:<10} | {inv.laboratorio:<15}")

        print("="*115)

    # --- MÉTODO: ACTUALIZAR ---
    def actualizar_investigador(self):
        self.listar_investigadores()
        if len(self.sistema_ia_investigadores) == 0:
            return

        print("\n--- 🔄 ACTUALIZAR DATOS ---")
        id_modificar = int(input("Ingresa el ID del investigador que deseas modificar: ")) - 1

        if 0 <= id_modificar < len(self.sistema_ia_investigadores):
            inv = self.sistema_ia_investigadores[id_modificar]
            print(f"\nModificando a investigador en la línea: {inv.linea}")

            inv.linea = input("Nueva línea de investigación: ")
            inv.publicaciones = input("Nuevo número de publicaciones: ")
            inv.universidad = input("Nueva universidad: ")
            inv.becas = input("Nuevas becas obtenidas: ")
            inv.sni = input("Nuevo nivel de SNI: ")
            inv.laboratorio = input("Nuevo laboratorio: ")

            print("\n🔄 ¡Datos actualizados correctamente!")
        else:
            print("❌ ID no válido. Volviendo al menú.")

    # --- MÉTODO: BORRAR ---
    def borrar_investigador(self):
        self.listar_investigadores()
        if len(self.sistema_ia_investigadores) == 0:
            return

        print("\n--- 🗑️ ELIMINAR INVESTIGADOR ---")
        id_borrar = int(input("Ingresa el ID del investigador que deseas dar de baja: ")) - 1

        if 0 <= id_borrar < len(self.sistema_ia_investigadores):
            eliminado = self.sistema_ia_investigadores.pop(id_borrar)
            print(f"\n🗑️ El investigador de '{eliminado.linea}' ha sido eliminado del sistema.")
        else:
            print("❌ ID no válido. Volviendo al menú.")

    # --- MÉTODO: MENÚ INTERACTIVO Y PRECARGA ---
    def iniciar_sistema(self):
        # Metemos los dos investigadores por defecto usando la lista del propio objeto ('self')
        self.sistema_ia_investigadores.append(Investigador("Visión por Computadora", "12", "UNAM", "Conahcyt", "Nivel I", "Lab de IA Alfa"))
        self.sistema_ia_investigadores.append(Investigador("Procesamiento de Lenguaje", "25", "IPN", "Google AI Grant", "Nivel II", "Lab Turing"))

        opcion = ""
        while opcion != "5":
            print("\n" + "•"*40)
            print("🤖 SISTEMA DE GESTIÓN: CIENTÍFICOS DE IA 🤖")
            print("•"*40)
            print("1. Agregar Investigador 🧪")
            print("2. Listar Investigadores (Ver Tabla) 📊")
            print("3. Actualizar Datos de Investigador 🔄")
            print("4. Borrar Investigador 🗑️")
            print("5. Salir del Sistema ❌")
            print("•"*40)

            opcion = input("Elige una opción (1-5): ")

            if opcion == "1":
                self.agregar_investigador()
            elif opcion == "2":
                self.listar_investigadores()
            elif opcion == "3":
                self.actualizar_investigador()
            elif opcion == "4":
                self.borrar_investigador()
            elif opcion == "5":
                print("\n🚀 Apagando sistema de Inteligencia Artificial... ¡Hasta la próxima, científicos!")
            else:
                print("⚠️ Opción no válida. Intenta de nuevo.")


# --- EJECUCIÓN DEL PROGRAMA ---
if __name__ == "__main__":
    # Creamos un objeto "sistema" usando la misma clase Investigador
    sistema = Investigador()
    # Ejecutamos el menú desde el objeto, usando 'self' de forma interna para todo
    sistema.iniciar_sistema()
```
